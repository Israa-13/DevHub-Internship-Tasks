## Load Dataset & Initial Inspection



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style='whitegrid')

iris = sns.load_dataset('iris')
print('Shape:', iris.shape)
iris.head()
```

## Exploratory Data Analysis

Distributions, pairwise relationships, boxplots, and correlations.

```python
# Descriptive stats
print(iris.describe())

# Pairplot
sns.pairplot(iris, hue='species', diag_kind='hist')
plt.suptitle('Iris pairwise relationships (by species)', y=1.02)
plt.show()
```

```python
# Histograms
numeric_cols = iris.select_dtypes(include=np.number).columns
iris[numeric_cols].hist(bins=12, figsize=(10,4))
plt.suptitle('Feature distributions')
plt.show()
```

```python
# Boxplots by species
plt.figure(figsize=(10,4))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(1, len(numeric_cols), i)
    sns.boxplot(x='species', y=col, data=iris)
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(6,5))
corr = iris[numeric_cols].corr()
sns.heatmap(corr, annot=True, center=0)
plt.title('Feature correlation matrix')
plt.show()
```

Summarize the key findings and actionable insights from EDA.
